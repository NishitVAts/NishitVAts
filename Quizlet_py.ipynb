Quizlet is a Machine learning based application which generates questions according to the user's current IQ.

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data collection and Processing

```python
movie_data = pd.read_csv('/content/movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
selected_feature = ['genres', 'keywords','tagline','cast','director']
print(selected_feature)
```

```python
#replacing the null value with null string

for feature in selected_feature:
  movie_data[feature] = movie_data[feature].fillna('')
```

```python
#combine all the selectedfeature
combined_feature = movie_data['genres'] +' '+ movie_data['keywords'] +' '+ movie_data['tagline'] +' '+ movie_data['cast'] +' '+ movie_data['director']
print(combined_feature)
```

```python
#converting text data into vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_feature)
print(feature_vectors)
```

Cosine Similarity

```python
#getting the similarity score
similarity = cosine_similarity(feature_vectors)
similarity.shape
```

```python
# getting the movie name from the user
movie_name = input("Enter your favourite movie name: ")
list_movies = movie_data['title'].tolist()
```

```python
#finding th eclosest match

closed_matches = difflib.get_close_matches(movie_name, list_movies)
print(closed_matches)
```

```python
closest_match = closed_matches[0]
index_of_movie = movie_data[movie_data.title == closest_match]['index'].values[0]
print(index_of_movie)
```

```python
#gettig the similar movies
similarity_score = list(enumerate(similarity[index_of_movie]))
similarity_score

```

```python
#sorting the movies based on higher similarity value
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies )
```

```python
# printing the name of similar movie
# index_of_movie = movie_data[movie_data.title == closest_match]['index'].values[0]
# recommended_movies_name = movie_data[movie_data.index == sorted_similar_movies]

print("Movies suggested for you /n")

i =1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_data[movie_data.index == index]['title'].values[0]
  if(i<30):
    print(i , '.', title_from_index)
    i+=1
```
